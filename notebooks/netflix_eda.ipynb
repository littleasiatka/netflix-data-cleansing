# üì∫ Netflix Titles Dataset ‚Äî Data Cleaning & EDA

This project explores and analyzes the Netflix titles dataset to uncover trends in content types, genres, release years and regional distributions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## üîç Load and Inspect Data

```python
df = pd.read_csv("../data/netflix_titles.csv")
df.head()
```

## üìä Basic Info and Missing Values

```python
df.info()
df.isnull().sum()
```

## üßπ Data Cleaning

```python
df = df.dropna(subset=['title'])
df['country'] = df['country'].fillna('Unknown')
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['date_added'].isna().sum()
```

## üìà EDA: Type of Content

```python
sns.countplot(data=df, x='type')
plt.title('Movie vs TV Show Count')
plt.show()
```

## üì∫ Top 10 Genres

```python
df['listed_in'].value_counts().head(10).plot(kind='barh', title='Top 10 Genres')
plt.show()
```

## üìÖ Content Added by Year

```python
df['year_added'].value_counts().sort_index().plot(kind='line', title='New Netflix Additions by Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## üåç Top 10 Countries by Content

```python
df['country'].value_counts().head(10).plot(kind='bar', title='Top 10 Countries Producing Netflix Content')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## üé¨ Release Year Distribution

```python
df['release_year'].plot(kind='hist', bins=30, title='Release Year Distribution', edgecolor='black')
plt.xlabel('Release Year')
plt.tight_layout()
plt.show()
```
